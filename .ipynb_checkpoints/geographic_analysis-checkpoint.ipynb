```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
# read in the data set
df = pd.read_csv('data_set/M_Landings_cleaned.csv')
df.head()
```

```python
# need to make a projection for all of earth to start
fig = plt.figure(figsize=(16,14), edgecolor='r') # define the size of the image as well as borders
m = Basemap(resolution='l')
m.bluemarble(scale=1)

# read in the latitude and longitude columns from the data set
lats = df['reclat']
lons = df['reclong']


x, y = m(lats, lons)
# create a scatter plot of meteorite landings
m.scatter(x, y, marker='D', color='r')
plt.show()
```
