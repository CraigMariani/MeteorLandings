```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
# practice map 
map = Basemap(projection='aeqd',
             lat_0=0, lon_0=0)

# map
map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='coral', lake_color='aqua')
map.drawcoastlines(color ='black')
plt.show()
```

```python
def draw_map(m, scale=0.2):
#     m.bluemarble(scale=scale)
    m.drawcoastlines(color='black')
```

```python
# need to make a projection for all of earth to start
fig = plt.figure(figsize=(16,14), edgecolor='r') # define the size of the image as well as borders
m = Basemap(resolution='l')
draw_map(m)
```

```python
# read in the data set
df = pd.read_csv('data_set/M_Landings_cleaned.csv')
df.head()
```

```python
# set x and y to reclat and reclong of our data set
x, y = m(df['reclat'], df['reclong'])
plt.scatter(x, y, color='red')
plt.show()
```
